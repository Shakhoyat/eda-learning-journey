### 3: Filling the Missing Values – Imputation

In this case, we will be filling the missing values with a certain number.

The possible ways to do this are:

- Filling the missing data with the mean or median value if it’s a numerical variable.
- Filling the missing data with mode if it’s a categorical value.
- Filling the numerical value with 0 or -999, or some other number that will not occur in the data.
  This can be done so that the machine can recognize that the data is not real or is different.
- Filling the categorical value with a new type for the missing values.

```python
df['Age'].mean()
```

```python
df['Age'].median()
```

```python
#fillna=fill the null attributes
#dropna=drop the null attributes
updated_df=df
updated_df['Age']=updated_df['Age'].fillna(updated_df['Age'].mean())
updated_df.info()
```

```python
updated_df2 = df
updated_df2['Age'] = updated_df2['Age'].fillna(updated_df2['Age'].median())
updated_df2.info()
```

Onek besi outliers thakle better to use median otherwise go for mean....


### 4. Forward & Backward Filling -Imputation 

```python
#ffill=forward fill
updated_df3 = df
updated_df3['Age'] = updated_df3['Age'].ffill()
updated_df3.info()
```

```python
#bfill=backward fill
updated_df4 = df
updated_df4['Age'] = updated_df4['Age'].bfill()
updated_df4.info()
```

# Feature Scaling
>Important for predictive analysis Data

1. Standardizations :
> Standardization is another scaling technique where the values are centered around the mean with a unit standard deviation .This means that the mean of the attribute becomes zero and the resultant distribution has a unit standard deviation.

```python
from sklearn.preprocessing import StandardScaler
standard_scaler = StandardScaler()
```

```python
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
updated_df5= pd.DataFrame(df, columns=['Age', 'Tenure'])
updated_df5.head()
```

```python
updated_df5.shape
```

```python
standardized_df=standard_scaler.fit_transform(updated_df5)
standardized_df
```

2 . Normalization :
>Normalization is another scaling technique where values are shifted and rescaled so that they end up ranging between 0 and 1 .It is also known as Min-Max scaling

```python
from sklearn.preprocessing import MinMaxScaler
minmax_scaler = MinMaxScaler()
```

```python
normalized_df = minmax_scaler.fit_transform(updated_df5)
normalized_df
```
