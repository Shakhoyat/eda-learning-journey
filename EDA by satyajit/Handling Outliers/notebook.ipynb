#### 1. 3-Sigma Technique(Standard Deviation) 

```python
#Function to detect outliers using 3-Sigma Technique on one-dimensional data
def detect_anomalies(data):
    anomalies=[]
    random_data_mean = statistics.mean(data) # miu
    random_data_std_dev = statistics.stdev(data) # sigma
    
    # Calculate the lower and upper bounds
    lower_bound = random_data_mean - 3 * random_data_std_dev # miu - 3*sigma
    upper_bound = random_data_mean + 3 * random_data_std_dev # miu + 3*sigma
    
    # Identify outliers
    for i in data:
        if i < lower_bound or i > upper_bound:
            anomalies.append(i)
    
    return anomalies
```
