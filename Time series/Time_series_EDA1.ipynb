# Assignment
1. Read Microsoft Data using Pandas Data Reader
2. Get Maximum price of the share from 2017 to 2022
3. which date is the highest price of the stock?
4. which is the date of the lowest price of the stock?

```python
import yfinance as yf
msft_data = yf.download('MSFT', start='2017-01-01', end='2022-12-31')
print(msft_data.head())
```

```python
max_price_msft = msft_data['High'].loc[:, 'MSFT'].max()
max_price_msft
```

```python
msft_data['High'].plot(figsize=(12,6), title='Microsoft Stock High Price Trend (2017-2022)')
plt.xlabel('Date')
plt.ylabel('High Price')
plt.grid(True)
```

```python
date_highest_price_msft = msft_data[
    msft_data['High'].loc[:, 'MSFT'] == max_price_msft
].index[0]

print("Highest High date:", date_highest_price_msft)
```

```python
min_price_msft = msft_data['Low'].loc[:, 'MSFT'].min()

min_price_msft
```

```python
msft_data['Low'].plot(figsize=(12,6), title='Microsoft Stock Low Price Trend (2017-2022)', color='red')
plt.xlabel('Date')
plt.ylabel('Low Price')
plt.grid(True)
```

```python
date_lowest_price_msft = msft_data[
    msft_data['Low'].loc[:, 'MSFT'] == min_price_msft
].index[0]

print("Lowest Low date:", date_lowest_price_msft)
```
